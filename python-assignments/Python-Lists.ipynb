##### Question 1
Write a program that accepts a list from user and print the alternate element of list.

```python
lst = input("Enter elements separated by space: ").split()
print("Alternate elements:")
for i in range(0, len(lst), 2):
    print(lst[i])
```

##### Question 2
Write a program that accepts a list from user. Your program should reverse the content of list and display it. Do not use reverse() method.

```python
lst = input("Enter elements separated by space: ").split()
reversed_lst = []
for i in range(len(lst)-1, -1, -1):
    reversed_lst.append(lst[i])
print("Reversed list:", reversed_lst)
```

##### Question 3
Find and display the largest number of a list without using built-in function max(). Your program should ask the user to input values in list from keyboard.

```python
lst = [int(x) for x in input("Enter numbers separated by space: ").split()]
largest = lst[0]
for num in lst:
    if num > largest:
        largest = num
print("Largest number:", largest)
```

##### Question 4
Write a program that rotates the element of a list so that the element at the first index moves to the second index, the element in the second index moves to the third index, etc., and the element in the last index moves to the first index.

```python
lst = input("Enter elements separated by space: ").split()
rotated = lst[1:] + lst[:1]
print("Rotated list:", rotated)
```

##### Question 6
Write a program that reads a string from the user containing a date in the form mm/dd/yyyy. It should print the date in the form March 12, 2021.

```python
months = ["January", "February", "March", "April", "May", "June",
          "July", "August", "September", "October", "November", "December"]
date = input("Enter date (mm/dd/yyyy): ")
mm, dd, yyyy = date.split("/")
month_name = months[int(mm)-1]
print(f"{month_name} {int(dd)}, {yyyy}")
```

##### Question 7
Write a program with a function that accepts a string from keyboard and create a new string after converting character of each word capitalized. For instance, if the sentence is "stop and smell the roses." the output should be "Stop And Smell The Roses"

```python
def capitalize_words(sentence):
    result = ""
    new_word = True
    for ch in sentence:
        if ch == " ":
            result += ch
            new_word = True
        elif new_word:
            result += ch.upper()
            new_word = False
        else:
            result += ch.lower()
    return result

s = input("Enter a sentence: ")
print("Capitalized:", capitalize_words(s))
```

##### Question 8
Find the sum of each row of matrix of size m x n. For example for the following matrix output will be like this :
2  11   7  12
5  2  9  15
8  3  10  42

Sum of row 1 = 32
Sum of row 2 = 31
Sum of row 3 = 63

```python
matrix = []
m = int(input("Enter number of rows: "))
n = int(input("Enter number of columns: "))
for i in range(m):
    row = [int(x) for x in input(f"Enter row {i+1} elements: ").split()]
    matrix.append(row)
for i in range(m):
    print("Sum of row", i+1, "=", sum(matrix[i]))
```

##### Question 9
Write a program to add two matrices of size n x m.

```python
m = int(input("Enter rows: "))
n = int(input("Enter columns: "))
print("Enter first matrix:")
A = [[int(x) for x in input().split()] for _ in range(m)]
print("Enter second matrix:")
B = [[int(x) for x in input().split()] for _ in range(m)]
C = [[A[i][j] + B[i][j] for j in range(n)] for i in range(m)]
print("Resultant matrix:")
for row in C:
    print(row)
```

##### Question 10
Write a program to multiply two matrices.

```python
m = int(input("Enter rows of first matrix: "))
n = int(input("Enter cols of first matrix: "))
A = [[int(x) for x in input().split()] for _ in range(m)]
p = int(input("Enter rows of second matrix: "))
q = int(input("Enter cols of second matrix: "))
B = [[int(x) for x in input().split()] for _ in range(p)]
if n != p:
    print("Matrix multiplication not possible")
else:
    result = [[0]*q for _ in range(m)]
    for i in range(m):
        for j in range(q):
            for k in range(n):
                result[i][j] += A[i][k]*B[k][j]
    print("Resultant matrix:")
    for row in result:
        print(row)
```
